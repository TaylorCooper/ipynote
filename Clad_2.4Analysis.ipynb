```python
###
### Useful functions
###

# Log file stuff
out4 = '2014.09.17_24_OpticsValidation_Analysis'
logPath = root + out4 + '.csv'
masterLogFile = open(logPath, 'wb')
mlog = csv.writer(masterLogFile, delimiter=',',quoting=csv.QUOTE_NONE)
headers = [
          'Filename',
          'Large_All (mean)',
          'Small_All (mean)',
          'Large_All (std)',
          'Small_All (std)',
          'Runtime (min)'
          ]
mlog.writerow(headers)

def printDiffs(series, threshold):
    """ For debugging, seeing where the break points are roughly
    """
    
    # Give me x - x.shift(3)
    for i,val in enumerate(series): 
        print i,val,
        if abs(val) > threshold: 
            print '  >', threshold,  

        print ''
        
def getInflectionPoints(diffArray, threshold, lookingForDrop):
    """ Takes a diffArray of ' Small' particle counts and a threshold and returns an array indicating which 
    dataset to count values in (high or low) and another array giving those indicies
    """

    shifted = diffArray
    
    #cycle = shifted[abs(shifted) > threshold] # threshold during
    
    negative = lookingForDrop
    points = [] # Points that indicate an inflection

    last = 0 # Variable to indicate last flip, enforces minimum width of high / low
    
    for i,val in enumerate(shifted):
        
        # If looking for negative and negative found 
        # and you've looked at more then 8 values since last find
        # and the current value is greater then the threshold 
        # record i+1 or i as an inflection point
        if negative and val < 0 and last > 8 and abs(val) > threshold: 
            negative = False
            last = 0
    #         print 'neg', val, cycle.index[i+1]
            if shifted[shifted.index[i+1]] < 0:
                points.append(shifted.index[i+1])
            else:
                points.append(shifted.index[i])

        # Same as above but now you are looking for positives
        elif not negative and val > 0 and last > 8 and abs(val) > threshold: 
            negative = True
            last = 0
    #         print 'pos', val, cycle.index[i+1]
            if shifted[shifted.index[i+1]] > 0:
                points.append(shifted.index[i+1])
            else:
                points.append(shifted.index[i])
        
        last +=1

    points.append(len(shifted)) # End point for dataset
    inflectionPoints = points
    high = lookingForDrop # True if high, False if Low
    j = 0
    isHigh = []
    
    for i in range(len(shifted)):
        
        #print i,inflectionPoints[j],high
        
        if high and i != inflectionPoints[j]:
            isHigh.append(True)
        elif high and i == inflectionPoints[j]:
            isHigh.append(True)
            j +=1
            high = False
        elif not high and i != inflectionPoints[j]:
            isHigh.append(False)
        elif not high and i == inflectionPoints[j]:
            isHigh.append(False)
            j +=1
            high = True

    return isHigh, inflectionPoints

def plotDF(df,infpts, name):
    
    timeElapsedMinutes = lambda x: x+1
    df['Minutes Elapsed']=df.index
    df['Minutes Elapsed']=df['Minutes Elapsed'].apply(timeElapsedMinutes)

    x=df['Minutes Elapsed']
    y1 = df[' Small'] # Great they have a leading space in there titles
    y2 = df[' Large']
    
    fig = plt.figure(figsize=(36,8))
    fig.suptitle(name, fontsize=12, fontweight='bold')
    ax = fig.add_subplot(121)
    
    for i in infpts:
        ax.axvline(i, color='k', ls='--', linewidth=0.2)
    ax.axis([0,max(x)+10, 0,max(y1)+5000])
    ax.set_ylabel('Small Particle Count', size=8)
    ax.set_xlabel('Time (min)', size=8)
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    ax.plot(x,y1,'b.')

    ax = fig.add_subplot(122)
    
    for i in infpts:
        ax.axvline(i, color='k', ls='--', linewidth=0.2)
    ax.axis([0,max(x)+10, 0,max(y2)+1000])
    ax.set_ylabel('Large Particle Count', size=8)
    ax.set_xlabel('Time (min)', size=8)
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    ax.plot(x,y2,'g.')
    
    savefig(root+name+'.svg') #, dpi=350, papertype='b10', orientation='portrait', bbox_inches='tight')
    show()
    clf()
```

```python
def plotDiffs(s, name):

    x= s.index
    y = s
    
    fig = plt.figure(figsize=(12,4))
    fig.suptitle(name, fontsize=12, fontweight='bold')
    ax = fig.add_subplot(111)

    ax.axis([0,max(x)*1.1, 0, max(y)*1.2])
    ax.set_ylabel(name, size=8)
    ax.set_xlabel('Time (min)', size=8)
#     ax.xaxis.set_ticks_position('none')
#     ax.yaxis.set_ticks_position('none')
    ax.plot(x,y,'bo')
   
    savefig(root+name+'.svg') #, dpi=350, papertype='b10', orientation='portrait', bbox_inches='tight')
    show()
    clf()

# Calculating stuff we really care about
S = abs(df41[' Small'] - df42[' Small'])
S = S.dropna()
L = abs(df41[' Large'] - df42[' Large'])
L = L.dropna()


# Plotting what we really care about
plotDiffs(S,'No_purifier-Small_Particle_Count')
plotDiffs(L,'No_purifier-Large_Particle_Count')

# Logging stuff we really care about
headersDiff = [
          'Filename1',
          'Filename2',
          'No_purifier-Small_Particle_Count(mean)',
          'No_purifier-Large_Particle_Count(mean)',
          'No_purifier-Small_Particle_Count(std)',
          'No_purifier-Large_Particle_Count(std)',
          ]

mlog.writerow([' '])
mlog.writerow([' '])

mlog.writerow(headersDiff)

row = [
          log41,
          log42,
          S.mean(),
          L.mean(),
          S.std(),
          L.std(),
          ]

mlog.writerow(row)
masterLogFile.close()
```
