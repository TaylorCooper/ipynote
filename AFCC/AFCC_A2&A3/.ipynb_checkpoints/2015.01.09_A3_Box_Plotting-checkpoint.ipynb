```python
import pandas as pd  # Pandas dataframes very cool for data analysis stuff
import numpy as np  # Numpy is math functionality 

import pylab as pl
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as tkr
import time, csv, os

%pylab inline  

path = "D:\\Dropbox\\Clients\\AFCC\\Projects\\Metal FSU\\A3 - Stamped Rapid-Prototyping\\Testing\\"
print 'Plotting from:', path
df = pd.read_csv(path+'20150112-ModifiedAnaylsisLog.csv', sep=',', header=0)
```

```python
###
###  Very important, must fill in table with NaN not - for invalid values
###

expts = ['AL01_A','AL01_B','AL01_C','B002_A','SS01_A','SS01_B','SS01_C','PC01_A',
         'PC01_B','PC01_C','PMMA01_A','PMMA01_B','PMMA01_C']

def plotterAFCC(d):
    """Plot a left, right side by side of some stuff
    """

    figName = d[0]
    yleft = d[1]
    tleft = d[2]
    ymax = d[3]
    col1 = d[4]
    
    fig = plt.figure(figsize=(12,4))

    # Create first plot
    ax = fig.add_subplot(111) # rows,cols,index 121 = 1row, 2cols, top left graph
    ax.set_xlim([0,14])  # use to adjust tick number to match what you need
    
    labels = ['']
    
#     y = df[df['Experiment']=='AL01_B'][col1]
#     print y
#     x = np.random.normal(0+1, 0.04, size=len(y))
#     print x

    for i,entry in enumerate(expts):
        y = df[df['Experiment']==entry][col1]
        labels.append(str(entry)+'  N='+str(len(y)))

        # Add some random "jitter" to the x-axis
        x = np.random.normal(i+1, 0.04, size=len(y))
        
        ax.plot(x, y, 'r.', alpha=0.5)
        ax.scatter(i+1, y.mean(), zorder=10000, alpha=0.7)

    # X tick nonsense
    ax.xaxis.set_major_locator(tkr.MaxNLocator(16))
    ax.set_xticklabels(labels)
    plt.xticks(rotation=70)
    for xticka in ax.xaxis.get_major_ticks():
        xticka.label1.set_fontsize(6)

    # Labels and scaling
    ax.set_ylabel(yleft, size=7)
    ax.set_xlabel('Experiment', size=7)
    ax.set_title(tleft, size=8)
    ax.set_ylim([0,ymax])

    savefig(figName, dpi=350, papertype='b10', orientation='portrait', bbox_inches='tight')#, format='pdf')
    show()
    clf()
```

```python
###
### AFCC ANODE AND CATHODE CA BY EXPERIMENT
###

figName = path + 'A3_AdvCA_Boxplots.png' 
yleft = 'Advancing Contact Angle (deg)'
tleft = 'Advancing Contact Angle (deg) vs. Experiment'
ymax = 100
col1 = 'MeanAdv(deg)'

variables = [figName,yleft,tleft,ymax,col1]             
plotterAFCC(variables)
```

```python
figName = path + 'A3_RecCA_Boxplots.png' 
yleft = 'Receding Contact Angle (deg)'
tleft = 'Receding Contact Angle (deg) vs. Experiment'
ymax = 100
col1 = 'MeanRec(deg)'

variables = [figName,yleft,tleft,ymax,col1]             
plotterAFCC(variables)
```
