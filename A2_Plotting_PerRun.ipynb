```python
import pandas as pd  # Pandas dataframes very cool for data analysis stuff
import numpy as np  # Numpy is math functionality 

import pylab as pl
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as tkr
import time, csv, os

%pylab inline  
```

```python
###
### Set run parameters and paths
###

# Run parameters
dataSet = 'AFCC'  # Change to INC\\ L\\ AFCC\\ to run different sets

#  Paths, assign them to the specific analysis folder for each experiment set
root = 'D:\\_Work\\0000-AFCC_MetalFSU_A2\\2014.08.28_Analysis\\'
dataPath = saveLoc = root + dataSet + '\\'
setMMRPath = dataPath + 'setMinMaxRegions' + dataSet + '.csv'
mlPath = dataPath + 'analysisLog' + dataSet + '_'

if dataSet == 'Test':
    dataPath = "D:\\_Work\\0000-AFCC_MetalFSU_A2\\2014.08.28_Analysis\\_Testing\\"
    saveLoc = "D:\\_Work\\0000-AFCC_MetalFSU_A2\\2014.08.28_Analysis\\_Testing\\"
    setMMRPath = "D:\\_Work\\0000-AFCC_MetalFSU_A2\\2014.08.28_Analysis\\_Testing\\setMinMaxRegionsTest.csv"
    mlPath = 'D:\\_Work\\0000-AFCC_MetalFSU_A2\\2014.08.28_Analysis\\_Testing\\analysisLogTest_'
    
```

```python
###
### General plots and analysis log
###

defaultRegions = False # Set to true when building setMMRPath

# Used to set adv / rec regions
df2 = pd.read_csv(setMMRPath, sep=',', header=0)

# Set logging file details
ts = str(time.time())[2:-3] # timestamp
mlPathFinal = mlPath + ts + '.csv'
masterLogFile = open(mlPathFinal, 'wb')
mlog = csv.writer(masterLogFile, delimiter=',',quoting=csv.QUOTE_NONE)
headers = [
          'Filename',
          'aMin(idx)',
          'aMax(idx)',
          'rMin(idx)',
          'rMax(idx)',
          'advTime_Mean(s)',
          'revTime_Mean(s)',
          'Dia_Init(mm)',
          'Dia_Max(mm)',
          # Left Side
          'lAdv_Mean(deg)',
          'lAdv_Std(deg)',
          'lRec_Mean(deg)',
          'lRec_Std(deg)',
          'lHyst_Mean(deg)',
          'lHyst_Std(deg)',
          # Right side
          'rAdv_Mean(deg)',
          'rAdv_Std(deg)',
          'rRec_Mean(deg)',
          'rRec_Std(deg)',
          'rHyst_Mean(deg)',
          'rHyst_Std(deg)',
          ]
mlog.writerow(headers)

# For loop that looks for all files ending in .txt in the file path above
for root, subFolders, files in os.walk(dataPath):
    
    for aFile in files:
        
        # Ignore stuff that's not a data csv
        if 'analysisLog' in aFile: continue
        if aFile[-9:] != '_data.csv': continue
            
        path = root + '\\' + aFile
        print "Analyzing: " + path
        
        # Get aMin aMax rMin rMax from setMMR\
        if defaultRegions:
            aMin = 0
            aMax = 112
            rMin = 143
            rMax = 173
        else:
            aMin = df2[aFile][0]
            aMax = aMin + df2[aFile][1]
            rMin = df2[aFile][2]
            rMax = rMin + df2[aFile][3]
            
        # Skip first row, header = first row after skip
        df = pd.read_csv(path, sep=',', skiprows=1, header=0)  

        # Calibrate data
        pxToMM = lambda x: 0.31*x/23.5 # Convert bubble diameter to mm from px
        framesToSeconds = lambda x: x/30

        # Reference the columns of your matrices
        x = df['frameNumber'].apply(framesToSeconds)
        dia = df['diameter'].apply(pxToMM)
        lCA = df['leftCA']
        rCA = df['rightCA']

        # Get values 
        xAdv = x[aMin:aMax].mean()
        xRec = x[rMin:rMax].mean()
        dIni = dia[0]
        dMax = dia.max()
        
        # Left edge
        lAdv_Mean = lCA[aMin:aMax].mean()
        lAdv_Std = lCA[aMin:aMax].std()

        # Make sure you are not referencing outside the dataseries with all NaN values dropped
        # Also make sure the user has not told you to ignore receding by setting rMin=0
        if rMin != 0 and rMin < len(lCA.dropna()):
            rMaxL = rMax if rMax < len(lCA) else len(lCA) # Check you don't go out of bounds
            lRec_Mean = lCA[rMin:rMaxL].mean()
            lRec_Std = lCA[rMin:rMaxL].std()
            lHyst_Mean = lAdv_Mean - lRec_Mean
            lHyst_Std = np.sqrt(lAdv_Std**2 + lRec_Std**2)
        else:
            lRec_Mean = lRec_Std = lHyst_Mean = lHyst_Std = '-'
        
        # Right edge
        rAdv_Mean = rCA[aMin:aMax].mean()
        rAdv_Std = rCA[aMin:aMax].std()
        if rMin != 0 and rMin < len(rCA.dropna()):
            rMaxR = rMax if rMax < len(rCA) else len(rCA)
            rRec_Mean = rCA[rMin:rMaxR].mean()
            rRec_Std = rCA[rMin:rMaxR].std()
            rHyst_Mean = rAdv_Mean - rRec_Mean
            rHyst_Std = np.sqrt(rAdv_Std**2 + rRec_Std**2)
        else:
            rRec_Mean = rRec_Std = rHyst_Mean = rHyst_Std = '-'

        # Append data to log file
        row = [
              aFile,
              aMin,
              aMax,
              rMin,
              rMax,
              xAdv,
              xRec,
              dIni,
              dMax,
              lAdv_Mean,
              lAdv_Std,
              lRec_Mean,
              lRec_Std,
              lHyst_Mean,
              lHyst_Std,
              rAdv_Mean,
              rAdv_Std,
              rRec_Mean,
              rRec_Std,
              rHyst_Mean,
              rHyst_Std
          ]
        mlog.writerow(row)
        masterLogFile.flush() # Flush data incase of crash

        # Start making a plot
        fig = plt.figure()
        ax = fig.add_subplot(111)
        ax2 = ax.twinx()

        # Plot Simple Things
        ax.plot(x,lCA,'r.',markersize=2)
        ax.plot(x,rCA,'b.',markersize=2)
        ax2.plot(x, dia, 'g')
        ax.plot(xAdv,lAdv_Mean,'ro', alpha=0.75)
        ax.plot(xAdv,rAdv_Mean, 'bo', alpha=0.75)
        if rMin != 0:
            if rRec_Mean != '-':
                ax.plot(xRec,rRec_Mean,'bo', alpha=0.75)
            if lRec_Mean != '-':
                ax.plot(xRec,lRec_Mean,'ro', alpha=0.75)

        # Set labeles
        ax.set_ylabel('Contact Angle (deg)', size=8)
        ax.set_xlabel('Time (s)', size=8)
        name = aFile[:-9]
        name = name + '  Adv=' + str(int((lAdv_Mean+rAdv_Mean)/2))
        if rMin != 0:
            if rRec_Mean != '-' and lRec_Mean != '-': 
                name = name + '  Rec=' + str(int((lRec_Mean+rRec_Mean)/2))
                name = name + '  Hyst=' + str(int((lHyst_Mean+rHyst_Mean)/2))
        ax.set_title(name, size=8)
        ax2.set_ylabel('Diameter (mm)', size=8)

        # Scale axis and set ticks
        ax.axis([0,max(x)+10, 0,110])
        ax2.axis([0,max(x)+10, 0,18])
        ax.xaxis.set_major_locator(tkr.MultipleLocator(20))
        ax.yaxis.set_major_locator(tkr.MultipleLocator(10))
        ax2.yaxis.set_major_locator(tkr.MultipleLocator(2))

        # Plot angle calculation boundaries
        ax.axvline(x[aMin]+1, color='k', ls='--', linewidth=0.65, alpha=0.5)
        ax.axvline(x[aMax], color='k', ls='--', linewidth=0.65, alpha=0.5)
        if rMin != 0:
            xMin = rMin if rMin < len(x) else len(x)-1
            xMax = rMax if rMax < len(x) else len(x)-1
            ax.axvline(x[xMin], color='k', ls='--', linewidth=0.65, alpha=0.5)
            ax.axvline(x[xMax], color='k', ls='--', linewidth=0.65, alpha=0.5)

        # Make legend
        p1 = plt.Rectangle((0, 0), 1, 1, fc='r')
        p2 = plt.Rectangle((0, 0), 1, 1, fc='b')
        p3 = plt.Rectangle((0, 0), 1, 1, fc='g')
        ax.legend([p1,p2,p3], ['Left CA','Right CA','Diameter'], 
               loc=1,prop={'size':8})

        # Save the output as a png
        savefig(saveLoc+aFile[:-9]+'.png', dpi=350, papertype='b10', orientation='portrait', bbox_inches='tight')
        show()
        clf()
            
masterLogFile.close()  # Close logfile so it can be read
```
