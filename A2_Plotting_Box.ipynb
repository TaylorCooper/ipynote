```python
import pandas as pd  # Pandas dataframes very cool for data analysis stuff
import numpy as np  # Numpy is math functionality 

import pylab as pl
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as tkr
import time, csv, os

%pylab inline  

path = "D:\\_Work\\0000-AFCC_MetalFSU_A2\\2014.08.28_Analysis\\"
print 'Plotting from:', path
df = pd.read_csv(path+'dataSummary.csv', sep=',', header=0)
```

```python
expts = ['L003B','INC002B','AFCCA3','3DS002B']

def plotter(d):
    """Plot a left, right side by side of some stuff
    """

    figName = d[0]
    yleft = d[1]
    yright = d[2]
    tleft = d[3]
    tright = d[4]
    ymax = d[5]
    col1 = d[6]
    col2 = d[7]    
    col3 = d[8]
    col4 = d[9]    
    
    fig = plt.figure(figsize=(12,4))

    # Create first plot
    ax = fig.add_subplot(121) # rows,cols,index 121 = 1row, 2cols, top left graph

    labels = ['']

    for i,entry in enumerate(expts):
        y = df[df['Experiment']==entry][col1]
        labels.append(entry+'  N='+str(len(y)))

        # Add some random "jitter" to the x-axis
        x = np.random.normal(i+1, 0.04, size=len(y))
        ax.plot(x, y, 'r.', alpha=0.5)
        ax.scatter(i+1, y.mean(), zorder=10000, alpha=0.7)

        # Error bars
        z = df[df['Experiment']==entry][col2]
        ax.errorbar(i+1, y.mean(), xerr=None, yerr=z.mean()*2, color='b', zorder=10000)

    # X tick nonsense
    ax.xaxis.set_major_locator(tkr.MaxNLocator(5))
    ax.set_xticklabels(labels)
    plt.xticks(rotation=70)
    for xticka in ax.xaxis.get_major_ticks():
        xticka.label1.set_fontsize(6)

    # Labels and scaling
    ax.set_ylabel(yleft, size=7)
    ax.set_xlabel('Experiment', size=7)
    ax.set_title(tleft, size=8)
    ax.set_ylim([0,ymax])


    # Create second plot
    ax = fig.add_subplot(122) # rows,cols,index 121 = 1row, 2cols, top left graph
    labels = ['']

    for i,entry in enumerate(expts):
        y = df[df['Experiment']==entry][col3]
        labels.append(entry+'  N='+str(len(y)))

        # Add some random "jitter" to the x-axis
        x = np.random.normal(i+1, 0.04, size=len(y))
        ax.plot(x, y, 'r.', alpha=0.5)
        ax.scatter(i+1, y.mean(), zorder=10000, alpha=0.7)    

        # Error bars
        z = df[df['Experiment']==entry][col4]
        ax.errorbar(i+1, y.mean(), xerr=None, yerr=z.mean()*2, color='b', zorder=10000)

    ax.xaxis.set_major_locator(tkr.MaxNLocator(5))
    ax.set_xticklabels(labels)
    plt.xticks(rotation=70)
    for xticka in ax.xaxis.get_major_ticks():
        xticka.label1.set_fontsize(6)
    ax.set_ylabel(yright, size=7)
    ax.set_xlabel('Experiment', size=7)
    ax.set_title(tright, size=8)
    ax.set_ylim([0,ymax])

    savefig(figName, dpi=350, papertype='b10', orientation='portrait', bbox_inches='tight')#, format='pdf')
    show()
```

```python
###
### HYSTERISIS BY EXPERIMENT
###

figName = path + 'Hyst_vs_Exp.png' 
yleft = 'Left Contact Angle Hystersis (deg)'
yright = 'Right Contact Angle Hystersis (deg)'
tleft = 'Left Contact Angle Hysteresis (deg) vs. Experiment'
tright = 'Right Contact Angle Hysteresis (deg) vs. Experiment'
ymax = 80
col1 = 'lHyst_Mean(deg)'
col2 = 'lHyst_Std(deg)'
col3 = 'rHyst_Mean(deg)'
col4 = 'rHyst_Std(deg)'

variables = [figName,yleft,yright,tleft,tright,ymax,col1,col2,col3,col4]             
plotter(variables)
```
