```python
import pandas as pd
import numpy as np
%pylab inline

path = 'C:\\Users\\tcooper\\Desktop\\SS Scoda Design\\S4+ Issues\\08_Hotspot Run Failure Analysis\\'
datalogPath = path + "02_Output and Analysis\\2014.07.03 Run Log V3.1.csv"
df = pd.read_csv(datalogPath, sep=',')
```

```python
### Assignment for which dataframe is being plotted
# Dataframe for all
print 'All:', len(df['Protocol'])
name = 'PctWt_BoxPlot_All' 

bp = df.boxplot(column='maxWt/Qnt', by='Pct Wt Mkr in DE Well', grid=False)
xticks(rotation=70)
bp.set_ylabel('maxWt/Qnt', size=8)
bp.set_xlabel('Pct Wt Mkr in DE Well', size=8)
bp.set_title('', size=11)
bp.set_ylim([0,6])

itemsX = []
for i in df['Pct Wt Mkr in DE Well']:
    if pd.notnull(i) and (i not in itemsX):
        itemsX.append(i)
        
itemsX.sort()
print itemsX
               
for i,entry in enumerate(itemsX):
    y = df[df['Pct Wt Mkr in DE Well']==entry]['maxWt/Qnt']
    print entry, len(y)
    # Add some random "jitter" to the x-axis
    x = np.random.normal(i+1, 0.04, size=len(y))
    plot(x, y, 'r.', alpha=0.5)
    scatter(i+1, y.mean(), zorder=10000)

l = Line2D([0,10],[1,1],color='k',linestyle='--')                                    
bp.add_line(l)    
    
savefig(name, dpi=350, papertype='b10', orientation='portrait', bbox_inches='tight')#, format='pdf')
show()

```
