```python
import pandas as pd
import numpy as np
%pylab inline

datalogPath = 'C:\\Users\\tcooper\\Desktop\\SS Scoda Design\\S4+ Issues\\08_Hotspot Run Failure Analysis\\Boxplots\\2014.06.11-outputCSV-V3.1 boxplotdelta-dCMdLEvsMkBe.csv'
df = pd.read_csv(datalogPath, sep=',')
print df['Wt  marker  behaviour'].tail()
```

```python
bp = df.boxplot(column='VF1  dLE_Wt  (pix)', by='Wt  marker  behaviour', grid=False)
xticks(rotation=70)
bp.set_ylabel('VF1  dLE_Wt  (pix)', size=8)
bp.set_xlabel('Mkr Behavior', size=8)
bp.set_title('', size=11)
bp.set_ylim([-300,50])

itemsX = []
for i in df['Wt  marker  behaviour']:
    if i not in itemsX:
        itemsX.append(i)
        
print itemsX
               
for i,entry in enumerate(itemsX):
    y = df[df['Wt  marker  behaviour']==entry]['VF1  dLE_Wt  (pix)']
    print entry, len(y)
    # Add some random "jitter" to the x-axis
    x = np.random.normal(i+1, 0.04, size=len(y))
    plot(x, y, 'r.', alpha=0.5)
    scatter(i+1, y.mean(), zorder=10000)

#Add lines for control points
ctrlPt = -80
axhline(ctrlPt,color='green')
text(4.6, ctrlPt, '-80 Adj +0.5C', horizontalalignment='left', rotation=0)
ctrlPt = -30
axhline(ctrlPt,color='green')
text(4.6, ctrlPt-5, '-30 Adj +0.8C', horizontalalignment='left', rotation=0)
ctrlPt = -5
axhline(ctrlPt,color='green')
text(4.6, ctrlPt, '-5 Adj +1.3C', horizontalalignment='left', rotation=0)  

savefig('BoxPlotDeltaWt', dpi=350, papertype='b10', orientation='portrait', bbox_inches='tight')#, format='pdf')
show()
```
